```python
from tkinter import * 
from tkinter import messagebox,ttk
import tkinter as tk
import time
from datetime import datetime 
from time import strftime 
import threading
import math
from pygame import mixer
```

```python
# TAMPILAN GUI
# Membuat latar GUI
root = Tk()
root.title('Python Project')
root.geometry("500x300") #ukuran
root.resizable(False,False) #membuat fixed position

notebook = ttk.Notebook(root)
tab1 = Frame(notebook)
tab2 = Frame(notebook)

notebook.add(tab1,text="Tab 1")
notebook.add(tab2,text="Tab 2")

# FUNCTION
# Fungsi perintah
def th():
    t1 = threading.Thread(target=a, args=())
    t2 = threading.Thread(target=aktif, args=())
    t1.start()
    t2.start()

# Fungsi Logika   
def a():
    w = inp_waktu.get()
    if w == "":
        msg = messagebox.showerror('Invalid data','Please enter valid time')
    else:
        Alarmtime = w
        CurrentTime = time.strftime("%H:%M:%S")
        while Alarmtime != CurrentTime:
            CurrentTime = time.strftime("%H:%M:%S")
            
        if Alarmtime == CurrentTime:
            mixer.init()
            mixer.music.load('tone.mp3')
            mixer.music.play()
            msg = messagebox.showinfo('It is time',f'{inp_pesan.get()}')
            if msg == 'ok':
                mixer.music.stop()

# fungsi Jam digital              
def waktu():
    string = strftime('%H:%M:%S %p') 
    jamm.config(text = string) 
    jamm.after(1000, waktu)
    
# Fungsi Jam Analog
def analog():
    hours = int(time.strftime("%H"))
    minutes = int(time.strftime("%M"))
    seconds = int(time.strftime("%S"))

    # updating seconds hand
    seconds_x = seconds_hand_len * math.sin(math.radians(seconds * 6)) + center_x
    seconds_y = -1 * seconds_hand_len * math.cos(math.radians(seconds * 6)) + center_y
    canvas.coords(seconds_hand, center_x, center_y, seconds_x, seconds_y)

    # updating minutes hand
    minutes_x = minutes_hand_len * math.sin(math.radians(minutes * 6)) + center_x
    minutes_y = -1 * minutes_hand_len * math.cos(math.radians(minutes * 6)) + center_y
    canvas.coords(minutes_hand, center_x, center_y, minutes_x, minutes_y)

    # updating hours hand
    hours_x = hours_hand_len * math.sin(math.radians(hours * 30)) + center_x
    hours_y = -1 * hours_hand_len * math.cos(math.radians(hours * 30)) + center_y
    canvas.coords(hours_hand, center_x, center_y, hours_x, hours_y)

    tab1.after(1000, analog)

# Fungsi status alarm
def aktif():
    w = inp_waktu.get()
    aktif=tk.Label(tab2,text='alarm '+str(w)+' dibuat',
                   font = ('calibri', 20), 
                   foreground = 'red')
    aktif.pack(anchor='center')


    



# TAMPILAN
# TAMPILAN HEADER(judul)
head =Label(tab1,text="ALARM ",font=('lucida bright',15,'bold'),
           background = 'white',
           foreground = 'blue')
head.pack(fill=X)

# TAMPILAN JAM DIGITAL
jamm = Label(tab1,font = ('calibri', 15), 
            background = 'white', 
            foreground = 'black')
jamm.place(x=330, y=70)
# Membuat tombol untuk menampilkan jam
button2 = Button(tab1,command = waktu,text = "Real Time",font=('comic sans',9))
button2.config(width =12)
button2.place(x=222,y=70)

# TAMPILAN JAM ANALOG
canvas = tk.Canvas(tab1, width=210, height=280,)
canvas.pack(side=LEFT)
# create background
bg = tk.PhotoImage(file='jam.png')
bg = bg.subsample(4) #resize, memperkecil ukuran
canvas.create_image(110, 120, image=bg)
## Membuat Jarum Jam
# titik tengah
center_x,center_y  = 110,120
# Menentukan Panjang Jarum
seconds_hand_len = 65
minutes_hand_len = 50
hours_hand_len = 30
# Membuat bentuk jarum
seconds_hand = canvas.create_line(150, 150, 150 + seconds_hand_len, 150 + seconds_hand_len, width=1.5, fill='red')
minutes_hand = canvas.create_line(150, 150, 150 + minutes_hand_len, 150 + minutes_hand_len, width=2, fill='blue')
hours_hand = canvas.create_line(150, 150, 150 + hours_hand_len, 150 + hours_hand_len, width=4, fill='black')

## INPUT WAKTU ALARM
# Label satuan
satuan = Label(tab1,text="    Jam   :   Menit   :   Detik",font=('comic sans',10))
satuan.place(x=300, y=110)
# label perintah
lbl_waktu = Label(tab1,text="Waktu Alarm",font=('comic sans',10))
lbl_waktu.place(x=220, y=130)
# tempat input waktu
inp_waktu = Entry(tab1,font=('comic sans',12),width=20)
inp_waktu.place(x=300, y=130)

## INPUT PESAN
# label perintah
lbl_pesan = Label(tab1,text="Catatan",font=('comic sans',10))
lbl_pesan.place(x=220,y=170)
# tempat input pesan
inp_pesan = Entry(tab1,font=('comic sans',12),width=20)
inp_pesan.place(x=300, y=170)
           

## membuat tombol mulai
start = Button(tab1,text="Start alarm",font=('comic sans',12),command=th)
start.config(width=28)
start.place(x=220, y=210)

notebook.pack(expand=True, fill="both")
analog()
mainloop()
```
